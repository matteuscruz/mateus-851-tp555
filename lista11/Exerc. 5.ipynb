```python
# Importando as bibliotecas necessárias:
import numpy as np
import seaborn as sns
from scipy.special import expit
import matplotlib.pyplot as plt
import matplotlib.colors as ListedColormap

#Importando as bibliotecas Scikit Learn:
from sklearn.linear_model import Perceptron
from sklearn.datasets import make_blobs

# Importando as métricas necessárias:
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Definindo o número de exemplos:
N = 1000

# Definindo os centros:
centers = [[1, 0], [0, 1]]

# Gerando o dataset:
X, y = make_blobs(n_samples=N, centers = centers, random_state=42)
```

**Plote os dados do conjunto de treinamento em relação às classes a que
pertencem. Ou seja, defina marcadores diferentes para identificar cada um das
classes na figura. Por exemplo, use círculos para denotar exemplos que
pertencem à classe 0 e quadrados para denotar exemplos que pertencem à
classe 1.**


```python
# Plotando o dataset:

idx0 = np.argwhere(y == 0)
idx1 = np.argwhere(y == 1)

fig = plt.figure(figsize=(5,5))
plt.plot(X[idx0,0], X[idx0,1], 'p', label='Class 0')
plt.plot(X[idx1,0], X[idx1,1], 'rs', label='Class 1')
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.legend()
plt.savefig('knn_for_two_classes.png', dpi=600)
plt.show()
```

**Instancie um objeto da classe Perceptron, use a linha abaixo para realizar a
instanciação.**

```python
# Instaânciando o perceptron:
per = Perceptron(random_state=42)

# Treinando o modelo:
per.fit(X, y)

# Fazendo a predição do modelo:
y_pred = per.predict(X)
```

**Em seguida treine o modelo, faça a predição com X e calcule a precisão deste
modelo como mostrado abaixo.**


```python
# Calculando a acurácia do modelo:
accuracy = accuracy_score(y, y_pred)

# Imprimindo o resultado encontrado:
print('Acurácia: {:.2f}%'.format(accuracy * 100))
```

**Plote a matriz de confusão e a figura com a fronteira de decisão para a
classificação com o Perceptron.**


```python

# Plot the confusion matrix.
fig = plt.figure(figsize=(6,6))
mat = confusion_matrix(y, y_pred)
sns.set()
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=['0','1'], yticklabels=['0','1'], cmap="Blues")
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.savefig('confusion_matrix_dt.png', dpi=600)
plt.show()
```

```python
# step size in the mesh
h = .002  

# Create color maps
cmap_light = ListedColormap(['green', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkgreen', 'c', 'darkblue'])

# Plot the decision boundary. 
plt.figure(1, figsize=(10, 10))

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = X.min() - .5, X.max() + .5
y_min, y_max = X.min() - .5, X.max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = per.predict(np.c_[xx.ravel(), yy.ravel()])

print(xx.shape)

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolors='k',s=20)
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.title('2-Class classification for DecisionTree Classifier\n')
plt.show()
```
