Geração de valores X & Y

Geração de valores em uma distribuição Gaussiana (Média = 0, Desvio padrão = 1)
Geração de valores uniformes

```python
# M = 1000

X = np.random.rand(1000).reshape((1000,1))
W = np.random.randn(1000).reshape((1000,1))
```

Geração do gráfico Y

```python
Y = 1.2 + 2.3 * X + 10 * W
```

Importação da biblioteca Matplot

```python
import matplotlib
import matplotlib.pyplot as plt
```

Plotanto os valores

```python
X = np.random.rand(1000).reshape((1000,1))
W = np.random.randn(1000).reshape((1000,1))

Y = 1.2 + 2.3 * X + 10 * W

fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
ax[0].plot(X,'p')
ax[1].plot(Y,'p')
```
