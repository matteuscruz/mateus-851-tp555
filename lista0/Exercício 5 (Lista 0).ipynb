Crie 2 vetores, x1 e x2, respectivamente, com valores uniformemente espaçados entre -10 e 11 com passos de 0.25 unidades. Em seguida crie o vetor y o qual é definido pela seguinte equação:

y = x1² + x2²

a) Plote um gráfico 3D com x1, x2 e y sendo plotados nos eixos x, y e z:

### Resolução

```python
#importanto as bibliotecas

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
# Criandos os dados requisitados

x1 = x2 = np.arange(-10,11,0.25)

x1, x2 = np.meshgrid(x1, x2)
y = (x1**2) + (x2**2)

#for itemA, itemB in zip(x1,x2):
#    itemA = itemA ** 2
#    itemB = itemB ** 2
#    y.append(itemA + itemB)
    
```

```python
#Plotando os valores: 

print(x1)
print('-----------------------------------------------------------')
print(x2)
print('-----------------------------------------------------------')
print(y)
```

```python
# Plotando os valores em 3D

fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x1,x2,y,cmap=cm.coolwarm,linewidth=0,antialiased=True)
```
