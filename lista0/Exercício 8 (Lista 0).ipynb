Modulador:

```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt

entrada = np.random.randint(0,4,1000)
symbols = np.zeros(1000, dtype='cfloat')

def modulador(val):
    for i in range(len(val)):
        if val[i] == 0:
            symbols[i] = symbols[i] + (-1/(2 ** 1/2)-1j/(2 ** 1/2))
        elif val[i] == 1:
            symbols[i] = symbols[i] + (-1/(2 ** 1/2)+1j/(2 ** 1/2))
        elif val[i] == 2:
            symbols[i] = symbols[i] + (+1/(2 ** 1/2)+1j/(2 ** 1/2))
        elif val[i] == 3:
            symbols[i] = symbols[i] + (+1/(2 ** 1/2)-1j/(2 ** 1/2))
            
modulador(entrada)

fig, ax = plt.subplots()

ax.scatter(symbols.real,symbols.imag)
```
