Importação das bibliotecas:

```python
import numpy as np

import math

import matplotlib

import matplotlib.pyplot as plt
```

Criação da função 'modulator':

```python
def modulador(val):
    for i in range(len(val)):
        if val[i] == 0:
            symbols[i] = symbols[i] + (-1/(2 ** 1/2)-1j/(2 ** 1/2))
        elif val[i] == 1:
            symbols[i] = symbols[i] + (-1/(2 ** 1/2)+1j/(2 ** 1/2))
        elif val[i] == 2:
            symbols[i] = symbols[i] + (+1/(2 ** 1/2)+1j/(2 ** 1/2))
        elif val[i] == 3:
            symbols[i] = symbols[i] + (+1/(2 ** 1/2)-1j/(2 ** 1/2))
```

Criação da matriz e teste de seus resultados:

```python
# Gerar valores aleatório de 0 a 3:

entrada = np.random.randint(0,4,1000)

# Gerar uma matriz vazia para preenchimento:

symbols = np.zeros(1000, dtype='cfloat')
```

Execução da função:

```python
# Executar a função apartir dos elementos gerados no vetor de entrada:

modulador(entrada)
```

Testanto cada entrada possível da função.

```python
# Ver quais valores estão adicionados na matriz de simbolos:

print(symbols)

# Ver quais valores foram gerados pelo vetor de entrada:

print(entrada)
```
