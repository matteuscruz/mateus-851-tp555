**3**. Crie um conjunto de dados de treinamento utilizando a função *make_blobs* comomostrado abaixo.

```python
# Importando as bibliotecas necessárias:
import numpy as np
import matplotlib.pyplot as plt

# Importando as bibliotecas Scikit learn:
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Gerando o dataset:
X, y = make_blobs(n_samples=150, n_features=2, centers=5,
                  cluster_std=1.0, shuffle=True,random_state=42)
```

**A**. Crie uma figura mostrando os dados de treinamento.

```python
# Plotagem do dataset:
fig = plt.figure(figsize=(10, 10))
plt.plot(X[:,0], X[:,1], 'b.', label='Dataset')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```

**B**. Após inspecionar a figura, decida quantos clusters devem ser utilizados com o algoritmo do k-Means.

*Devem ser usados 4 clusters* 

**C**. Instancie um objeto da classe KMeans da biblioteca SciKit-Learn.

```python
# Instânciando o KMeans:
km = KMeans(n_clusters=4)
```

**D**. Treine o modelo e imprima os centróides ótimos.

```python
# Treinando o modelo:
km.fit(X)

# Imprimindo os centroides:
# Imprimindo os resultados:
for i in range(0,3):    
    print('Centroid %d: (%1.2f, %1.2f)' % (i,km.cluster_centers_[i][0],
                                           km.cluster_centers_[i][1]))
```

**E**. Quantas iterações foram necessárias para se treinar o modelo?

```python
# Imprimindo o número de iterações:
print('Foram necessárias {} iterações'.format(km.n_iter_))
```

**E**. Crie uma figura com os dados de treinamento indicando através de cores ou marcadores diferentes à que clusters cada um deles pertence além de mostraros centróides encontrados pelo k-Means.

```python
# Plotagem do dataset e os centróides:
fig = plt.figure(figsize=(12,12))
plt.plot(X[:,0], X[:,1], 'b.', label='dataset')
plt.plot(km.cluster_centers_[:,0], km.cluster_centers_[:,1], 'rx', label='KMeans')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.show()
```
